## Statistical Inference

* Population is the pool of individuals from which a statistical sample is drawn for a study. 
* Sample sampling is a subset of individuals from within a statistical population to estimate characteristics of the whole population. 

```python
# We can check these figure using cumulative density functions
from scipy import stats
mu_estim = stats.norm(170, 0.63)

# use the cdf to find the probabilities associated with height values
print('% confidence interval = ', mu_estim.cdf(171.2) - mu_estim.cdf(168.7))
```

2 - <b>Test the hypothesis:</b>

* If p-value > α, fail to reject the null hypothesis. This doesn't mean we accept H0.
